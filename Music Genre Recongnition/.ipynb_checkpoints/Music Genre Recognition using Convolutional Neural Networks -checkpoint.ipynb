```python
# Load Dependencies
import os
from matplotlib import pyplot as plt
import tensorflow as tf 
import tensorflow_io as tfio
```

```python
# Define Paths to Files 
blues = os.path.join('genres_original/blues/blues.00000.wav')
classical = os.path.join('genres_original/classical/classical.00000.wav')
country = os.path.join('genres_original/country/country.00000.wav')
disco = os.path.join('genres_original/disco/disco.00000.wav')
hiphop = os.path.join('genres_original/hiphop/hiphop.00000.wav')
jazz = os.path.join('genres_original/jazz/jazz.00000.wav')
metal = os.path.join('genres_original/metal/metal.00000.wav')
pop = os.path.join('genres_original/pop/pop.00000.wav')
reggae = os.path.join('genres_original/reggae/reggae.00000.wav')
rock = os.path.join('genres_original/rock/rock.00000.wav')
```

```python
# Build Dataloading Function 
def load_wav_16k_mono(filename):
    # Load encoded wav file
    file_contents = tf.io.read_file(filename)
    # Decode wav (tensors by channels) 
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    # Removes trailing axis
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    # Goes from 44100Hz to 16000hz - amplitude of the audio signal
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    return wav
```

```python
# call lload_wav_16k_mono
blues_f = load_wav_16k_mono(blues)
classical_f = load_wav_16k_mono(classical)
country_f = load_wav_16k_mono(country)
disco_f = load_wav_16k_mono(disco)
hiphop_f = load_wav_16k_mono(hiphop)
jazz_f = load_wav_16k_mono(jazz)
metal_f = load_wav_16k_mono(metal)
pop_f = load_wav_16k_mono(pop)
reggae_f = load_wav_16k_mono(reggae)
rock_f = load_wav_16k_mono(rock)
```

```python
# Plot Wave
plt.plot(blues_f)
plt.plot(classical_f)
plt.plot(country_f)
plt.plot(disco_f)
plt.plot(hiphop_f)
plt.plot(jazz_f)
plt.plot(metal_f)
plt.plot(pop_f)
plt.plot(reggae_f)
plt.plot(rock_f)
```
