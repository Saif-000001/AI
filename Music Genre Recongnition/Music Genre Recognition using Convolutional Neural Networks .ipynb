```python
import os
import subprocess
from matplotlib import pyplot as plt 
```

```python
DATASET_FOLDER = "dataset_folder"
SEGMENT_DURATION = 3
```

```python
def get_total_duration(filename):
    result = subprocess.run(
        ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", filename],
        capture_output=True,
        text=True
    )
    return float(result.stdout)
```

```python
def load_wav():
    for root, dirs, files in os.walk(DATASET_FOLDER):
        for filename in files:
            if filename.endswith(".wav"):
                filepath = os.path.join(root, filename)
                total_duration = get_total_duration(filepath)
                num_segments = int((total_duration + SEGMENT_DURATION - 1) / SEGMENT_DURATION)

                for i in range(num_segments):
                    start_time = i * SEGMENT_DURATION
                    output_file = os.path.join(root, f"{os.path.splitext(filename)[0]}_segment{i}.wav")

                    result = subprocess.run(
                        ["ffmpeg", "-y", "-loglevel", "error", "-i", filepath, "-ss", str(start_time), "-t", str(SEGMENT_DURATION), "-c", "copy", output_file]
                    )
                    print(result)
    
              
```

```python
load_wav()
```
